```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('heart_failure.csv')
data.head()

data = data.rename(columns={'DEATH_EVENT':'Target'})
data.head()
```

```python
data.isna().any()
```

```python
print(data.nunique())
print(data.shape)
```

```python
data.info()
```

```python
corr = data.corr()

fig = plt.figure(figsize=(15, 10))
sns.heatmap(corr, annot=True)
```

```python
#plot for smoking by sex

sns.countplot(data=data, x='smoking', hue='sex')

print(data.groupby('sex')['smoking'].value_counts())
```

```python
#plots counting Target and Sex

figs, (ax1, ax2) = plt.subplots(1,2,sharey=True)
figs = plt.figure(figsize=(10,5))
sns.countplot(ax=ax1, data=data, x='Target')
sns.countplot(ax=ax2, data=data, x='sex')

print(data.Target.value_counts())
```

#should oversample to level out target class between no event and event occuring

```python
#plot serum_creatine

fig = plt.figure(figsize=(10,5))
plt.scatter(x=data.serum_creatinine, y=data.Target, c=data.sex)
```

```python
#plot time by target 
fig = plt.figure(figsize=(10,5))
plt.scatter(x=data.time, y=data.Target, c=data.sex)
```

It does look like that the greater time that passes for the follow up to occur makes it more likely that any given patient
will not suffer from heart failure

```python
min_age = data['age'].agg(min)
max_age = data['age'].agg(max)
mean_age = data['age'].agg(np.mean)

print(min_age)
print(max_age)
print(mean_age)
```

```python
#plot age and serum_creatinine 
#look for range of age in graph for binning and further investigation
fig, (ax, ax1, ax2) = plt.subplots(3,1, figsize=(10,15))
ax1.scatter(x=data.serum_creatinine, y=data.age, c=data.Target)
ax1.legend(data.Target, loc='best')

age_sorted = data.age.sort_values()
age_binned = pd.cut(x=data.age, bins=4)

age_df = data[['age', 'Target']]
age_df = pd.DataFrame(age_df)
age_df['age_binned'] = age_binned

ax = sns.countplot(data=age_df, x='age_binned', hue='Target', ax=ax)

ax2.scatter(x=data.ejection_fraction, y=data.Target, c=data.smoking)
ax2.legend(data.Target, loc='best')
```

```python
columns = data.columns

labels = []
cont_vars = list()

for col in columns:
    if data[col].nunique() > 2:
        cont_vars.append(data[col])
        labels.append(col)
        
cont_df = pd.DataFrame(data=cont_vars)
cont_df.reset_index()
cont_df = cont_df.pivot_table(columns=cont_df.index)
cont_df['Target'] = data.Target

sns.pairplot(data=cont_df, hue='Target')
```
